```python
from PIL import Image
import os

# Đường dẫn tới thư mục chứa ảnh
train_image_dir = 'C:/Users/Modern/Documents/Traffic Sign Recog/TSD-1/images/train'

# Đường dẫn tới thư mục chứa tệp label
train_label_dir = 'C:/Users/Modern/Documents/Traffic Sign Recog/TSD-1/labels/train'

# Lấy danh sách tên tệp ảnh
image_file_names = os.listdir(train_image_dir)

for image_filename in image_file_names:
    # Tạo đường dẫn tới tệp ảnh và tệp label tương ứng
    image_path = os.path.join(train_image_dir, image_filename)
    label_filename = os.path.splitext(image_filename)[0] + '.txt'
    label_path = os.path.join(train_label_dir, label_filename)
    
    # Đọc tọa độ từ tệp label
    with open(label_path, 'r') as label_file:
        label_info = label_file.read().strip().split()
        label = [float(val) for val in label_info]
    
    # Load ảnh
    image = Image.open(image_path)
    
    # Tính toán tọa độ của khung cần cắt
    image_width, image_height = image.size
    x1 = int((label[1] - label[3] / 2) * image_width)
    y1 = int((label[2] - label[4] / 2) * image_height)
    x2 = int((label[1] + label[3] / 2) * image_width)
    y2 = int((label[2] + label[4] / 2) * image_height)
    
    # Cắt ảnh theo tọa độ
    cropped_image = image.crop((x1, y1, x2, y2))
    
    # Lưu ảnh đã cắt vào thư mục cắt
    cropped_image_path = os.path.join('C:/Users/Modern/Documents/Traffic Sign Recog/TSD-1/cropped_train', image_filename)
    cropped_image.save(cropped_image_path)
```

```python
import os
import cv2

# Đường dẫn tới thư mục chứa ảnh
image_dir = 'C:/Users/Modern/Documents/Traffic Sign Recog/TSD-1/cropped_train'

# Kích thước mà bạn muốn chúng có
desired_width = 91
desired_height = 92

# Đường dẫn tới thư mục lưu ảnh sau khi thêm border
output_dir = 'C:/Users/Modern/Documents/Traffic Sign Recog/TSD-1/border0_cropped_train'

# Lặp qua tất cả các tệp trong thư mục ảnh
for image_filename in os.listdir(image_dir):
    image_path = os.path.join(image_dir, image_filename)
    
    # Đọc ảnh bằng OpenCV
    image = cv2.imread(image_path)
    
    # Lấy kích thước ảnh gốc
    original_height, original_width = image.shape[:2]
    
    # Tính toán kích thước của border
    top_border = (desired_height - original_height) // 2
    bottom_border = desired_height - original_height - top_border
    left_border = (desired_width - original_width) // 2
    right_border = desired_width - original_width - left_border
    
    # Thêm border 0 vào ảnh
    image_with_border = cv2.copyMakeBorder(image, top_border, bottom_border, left_border, right_border, cv2.BORDER_CONSTANT, value=[0, 0, 0])
    
    # Lưu ảnh mới vào thư mục đầu ra
    output_path = os.path.join(output_dir, image_filename)
    cv2.imwrite(output_path, image_with_border)
```

```python
import cv2
import numpy as np
import os
import pandas as pd

# Đường dẫn tới thư mục chứa ảnh và label
image_dir = 'C:/Users/Modern/Documents/Traffic Sign Recog/TSD-1/border0_cropped_train'
label_dir = 'C:/Users/Modern/Documents/Traffic Sign Recog/TSD-1/labels/train'

# Số loại ảnh
num_classes = 6

# Khởi tạo danh sách DataFrame để lưu image_array và image_filename theo loại
dataframes_by_label = [pd.DataFrame(columns=['image_array', 'image_filename']) for _ in range(num_classes)]

# Lặp qua tất cả các tệp trong thư mục ảnh
for image_filename in os.listdir(image_dir):
    image_path = os.path.join(image_dir, image_filename)
    label_filename = os.path.splitext(image_filename)[0] + '.txt'
    label_path = os.path.join(label_dir, label_filename)

    # Kiểm tra xem tệp label có tồn tại hay không
    if os.path.exists(label_path):
        # Đọc nội dung tệp label
        with open(label_path, 'r') as label_file:
            label_info = label_file.read().strip().split()
            label_type = int(label_info[0])

        # Đọc ảnh và chuyển thành ma trận numpy
        image = cv2.imread(image_path)
        image_array = np.array(image)

        # Tạo một DataFrame mới chứa thông tin của image_array và image_filename
        df = pd.DataFrame({'image_array': [image_array], 'image_filename': [image_filename]})

        # Thêm DataFrame mới vào danh sách DataFrame tương ứng với label
        dataframes_by_label[label_type] = pd.concat([dataframes_by_label[label_type], df], ignore_index=True)

# Bây giờ dataframes_by_label chứa danh sách DataFrame theo từng loại ảnh
```
