```python
import os
```

```python
dir_list = r'Z:\train_label_new\run_8\labels'
```

```python
item_list = os.listdir(os.path.join(dir_list))
```

```python
for item in item_list:
    file_path = os.path.join(dir_list, item)  # Thay đổi thành đường dẫn của tệp .txt bạn muốn kiểm tra và xóa

    # Mở tệp văn bản trong chế độ đọc (read)
    with open(file_path, "r") as file:
        content = file.read()

    # Kiểm tra xem nội dung của tệp có rỗng không
    if not content:
        # Nếu tệp rỗng, bạn có thể xóa nó bằng cách sử dụng hàm os.remove
        import os
        os.remove(file_path)
        print(f"File '{file_path}' has been deleted because it was empty.")
    else:
        print(f"File '{file_path}' is not empty.")
```
