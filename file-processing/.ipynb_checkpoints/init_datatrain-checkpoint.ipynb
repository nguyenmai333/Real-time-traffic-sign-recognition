### Get Images From Labels

```python
import os
import shutil
```

```python
data_dir = r'Z:\train_label_new\run_7'
dir_labels = r"Z:\train_label_new\run_7\labels"
dir_images_new = r"Z:\train_label_new\run_7\images"
dir_images = r"Z:\init\frame-s"
os.makedirs(dir_images_new, exist_ok=True)
```

```python
list_labels = os.listdir(dir_labels)
list_images = os.listdir(dir_images)
```

```python
name_file = []
for label in list_labels:
    label = os.path.splitext(label)[0]
    name_file.append(label)
```

```python
name_file
```

```python
for file in name_file:
    source_dir = os.path.join(dir_images, file)
    source_dir += '.jpg'
    print(source_dir)
    shutil.copy(source_dir, dir_images_new)
```

# Rename File with index start 0001

### file .jpg

```python
#test folder "0" 20 limited speed
folder_path = dir_images_new
file_list = os.listdir(folder_path)

for i, file_name in enumerate(file_list, 1):
    
    old_path = os.path.join(folder_path, file_name)
    new_name = f"{i:03d}.jpg"
    new_path = os.path.join(folder_path, new_name)
    os.rename(old_path, new_path)
    
    print(f"Đã đổi tên {file_name} thành {new_name}")
```

### file .txt

```python
#test folder "0" 20 limited speed
folder_path = dir_labels
file_list = os.listdir(folder_path)

for i, file_name in enumerate(file_list, 1):
    
    old_path = os.path.join(folder_path, file_name)
    new_name = f"{i:03d}.txt"
    new_path = os.path.join(folder_path, new_name)
    os.rename(old_path, new_path)
    
    print(f"Đã đổi tên {file_name} thành {new_name}")
```

# Split_Train_Validation

```python
# Đường dẫn đến thư mục chứa ảnh và nhãn của tập huấn luyện và tập kiểm tra
train_dir = os.path.join(data_dir, 'images\\train')
val_dir = os.path.join(data_dir, 'images\\validation')
label_train_dir = os.path.join(data_dir, 'labels\\train')
label_val_dir = os.path.join(data_dir, 'labels\\validation')
```

```python
#create train,val,label folder
os.makedirs(train_dir, exist_ok=True)
os.makedirs(val_dir, exist_ok=True)
os.makedirs(label_train_dir, exist_ok=True)
os.makedirs(label_val_dir, exist_ok=True)

validation_ratio = 0.2 #random on 0.2per
image_files = os.listdir(os.path.join(data_dir, 'images'))
num_validation = int(validation_ratio * len(image_files))
validation_images = image_files[:num_validation]
```

```python
for image_file in validation_images:
    label_file = image_file.replace('.jpg', '.txt')  # Điều này giả sử nhãn có định dạng .txt
    shutil.move(os.path.join(data_dir, 'images', image_file), os.path.join(val_dir, image_file))
    shutil.move(os.path.join(data_dir, 'labels', label_file), os.path.join(label_val_dir, label_file))
```

```python
remaining_image_files = os.listdir(os.path.join(data_dir, 'images'))

for image_file in remaining_image_files:
    label_file = image_file.replace('.jpg', '.txt')  # Điều này giả sử nhãn có định dạng .txt
    shutil.move(os.path.join(data_dir, 'images', image_file), os.path.join(train_dir, image_file))
    shutil.move(os.path.join(data_dir, 'labels', label_file), os.path.join(label_train_dir, label_file))
```
